```python
df = pd.read_excel(path_data + "IOW - 3 - Cleaned Text.xlsx")
q = pd.read_excel(path_data + "Question_Index.xlsx")
```

## 1 SERVICE Replace misssing data

```python
(df[q[q.source == "service"].renamed.tolist()] == "DATA_EXPIRED").mean().apply("{:.2%}".format)
```

```python
df.loc[:, q[q.source == "service"].renamed.tolist()] = df.loc[:, q[q.source == "service"].renamed.tolist()].replace("DATA_EXPIRED", np.nan)
```

## 2 SPECIFIC Enrich PERSON_gender 

```python
def transfer_gender(row):

    if row.PERSON_relationship == "Wife or girlfriend":
        return "Female"

    elif row.PERSON_relationship == "Sister":
        return "Female"

    elif row.PERSON_relationship == "Husband or boyfriend":
        return "Male"

    elif row.PERSON_relationship == "Brother":
        return "Male"

    else:
        return row.PERSON_gender

df.PERSON_gender = df[["PERSON_relationship", "PERSON_gender"]].apply(transfer_gender, axis=1)
```

## 3 Nominal Columns

```python
nominal_columns = q[q.question_type == "nominal"].renamed.tolist()

pd.concat([
    df[column].value_counts().reset_index() for column in nominal_columns
],axis=1).fillna("-").to_excel(path_data + "ColumnInfo - Nominal.xlsx")
```
