```python
#Imports and data load
import warnings
warnings.filterwarnings('ignore')

import os
import sys
module_paths = [os.path.abspath(os.path.join('../py_scripts/')), os.path.abspath(os.path.join('../libraries/'))]
for module_path in module_paths:
    if module_path not in sys.path:
        sys.path.append(module_path)

import pandas as pd
import numpy as np
import seaborn as sns
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/GlasgowCleaned.csv")
matrix_distance =  pd.read_csv("../data/matrix/dat_distance.csv", index_col="word1")
```

### Gender

```python
ax = sns.barplot(df.sex.value_counts().reset_index(), x="index", y="sex")
ax.bar_label(ax.containers[0])
plt.xlabel("Sex")
plt.ylabel("Number of Participants")
```

### Age

```python
ax = sns.histplot(df.age)
plt.xlabel("Age")
plt.ylabel("Number of Participants")
```

```python
ax = sns.barplot(df.age_group.value_counts().reset_index(), x="index", y="age_group")
ax.bar_label(ax.containers[0])
plt.xlabel("Age group")
plt.ylabel("Number of Participants")
```

### Employment Status

```python
ax = sns.barplot(df.employment_status.value_counts().reset_index(), x="index", y="employment_status")
ax.bar_label(ax.containers[0])
plt.xlabel("Employment status")
plt.ylabel("Number of Participants")
```

### Country in the UK

```python
ax = sns.barplot(df.country.value_counts().reset_index(), x="country",y="index")
ax.bar_label(ax.containers[0])
plt.xlabel("Country of Residence")
plt.ylabel("Number of Participants")
```
